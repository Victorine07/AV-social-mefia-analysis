```python
# importing required modules 
import PyPDF2 
import pandas as pd
```

```python
pdfname = "Australia 1"
```

```python
# creating a pdf file object 
pdfFileObj = open(pdfname + '.pdf', 'rb') 
    
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
```

```python
# printing number of pages in pdf file 
print(pdfReader.numPages) 
numPg = pdfReader.numPages
x = 0

df = pd.DataFrame()
fullText = []

# creating a page object 
while x < numPg:
    pageObj = pdfReader.getPage(x) 
    text = pageObj.extractText()
    print(text)
    fullText.append(text)
    x = x + 1

print(fullText)
df['full text'] = fullText
df.to_csv(pdfname + '.csv', index = False, encoding='utf-8-sig')
    
# # extracting text from page 
# print(pageObj.extractText()) 
    
# closing the pdf file object 
pdfFileObj.close()
```

```python
# LOWERCASE
df = pd.read_csv(pdfname + ".csv", encoding='utf-8-sig')
df['full text'] = df['full text'].str.lower()
df.to_csv(pdfname + ".csv", index = False, encoding='utf-8-sig')
print(df['full text'])
```

```python
# DELETING NUMBERS

df = pd.read_csv(pdfname + ".csv", encoding='utf-8-sig')
df['full text'] = df['full text'].str.replace('\d+', '')
df.to_csv(pdfname + ".csv", index = False, encoding='utf-8-sig')
```

```python
df = pd.read_csv(pdfname + ".csv", encoding='utf-8-sig')
df['full text'] = df['full text'].str.replace('[^\w\s]',' ')
#df['hashtags'] = df['hashtags'].str.replace('[^\w\s]','')
df.to_csv(pdfname + ".csv", index = False, encoding='utf-8-sig')
```

```python
# WHITESPACE
df = pd.read_csv(pdfname + ".csv", encoding='utf-8-sig', skipinitialspace = True)
df = df.replace(r'\n',' ', regex=True)
df['full text'] = df['full text'].replace(r'\s+', ' ', regex=True)
df.dropna(subset = ['full text'], inplace=True)
df.to_csv(pdfname + ".csv", index = False, encoding='utf-8-sig')
print(df)
```

```python
df = pd.read_csv(pdfname + ".csv", encoding='utf-8-sig')
text = ['national enforcement guidelines for automated vehicles november','national enforcement guidelines for automated november','page of','references', 'appendix', 'table of contents', 'introduction']
df['full text'] = df['full text'].str.replace('|'.join(text), '')
df.to_csv(pdfname + ".csv", index = False, encoding='utf-8-sig')
print(df)
```
